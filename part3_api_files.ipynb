## TASK 1 — File Read & Write

```python
# Writing initial notes to file

notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes."
]

try:
    # Open file in write mode
    with open("python_notes.txt", "w", encoding="utf-8") as file:
        for line in notes:
            file.write(line + "\n")

    print("File written successfully.")

except Exception as e:
    print("Error writing file:", e)
```

```python
# Adding extra lines to same file

extra_lines = [
    "Topic 6: Functions help reuse code.",
    "Topic 7: APIs allow communication between systems."
]

try:
    # Open file in append mode
    with open("python_notes.txt", "a", encoding="utf-8") as file:
        for line in extra_lines:
            file.write(line + "\n")

    print("Lines appended successfully.")

except Exception as e:
    print("Error appending file:", e)
```

```python
# Reading file and printing numbered lines

print("\nReading file contents:\n")

try:
    with open("python_notes.txt", "r", encoding="utf-8") as file:
        lines = file.readlines()

    for i in range(len(lines)):
        print(f"{i+1}. {lines[i].strip()}")

except Exception as e:
    print("Error reading file:", e)
```

```python
# Counting total lines

try:
    print("\nTotal number of lines:", len(lines))

except Exception as e:
    print("Error counting lines:", e)
```

```python
# Searching for keyword in file

try:
    keyword = input("\nEnter keyword to search: ").lower()

    found = False

    for line in lines:
        if keyword in line.lower():
            print(line.strip())
            found = True

    if not found:
        print("No matching lines found.")

except Exception as e:
    print("Error during search:", e)
```

## Task 2 — API Integration

```python
# Install requests
!pip install requests

import requests
```

```python
print("\nFetching products...\n")

try:
    url = "https://dummyjson.com/products?limit=20"

    response = requests.get(url, timeout=5)

    data = response.json()
    products = data["products"]

    # Print table header
    print("ID | Title                 | Category        | Price    | Rating")
    print("-" * 70)

    for p in products:
        print(f"{p['id']}  | {p['title'][:20]:<20} | {p['category']:<14} | ${p['price']:<8} | {p['rating']}")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print("Error:", e)
```

```python
print("\nFiltered Products (Rating >= 4.5):\n")

try:
    # Filter products
    filtered = []

    for p in products:
        if p["rating"] >= 4.5:
            filtered.append(p)

    # Sort by price (descending)
    filtered.sort(key=lambda x: x["price"], reverse=True)

    # Print result
    for p in filtered:
        print(f"{p['title']} | ${p['price']} | Rating: {p['rating']}")

except Exception as e:
    print("Error during filtering/sorting:", e)
```

```python
print("\nFetching laptops category...\n")

try:
    url = "https://dummyjson.com/products/category/laptops"

    response = requests.get(url, timeout=5)

    data = response.json()
    laptops = data["products"]

    for p in laptops:
        print(f"{p['title']} | ${p['price']}")

except requests.exceptions.ConnectionError:
    print("Connection failed.")

except requests.exceptions.Timeout:
    print("Request timed out.")

except Exception as e:
    print("Error:", e)
```

```python
print("\nSending POST request...\n")

try:
    url = "https://dummyjson.com/products/add"

    new_product = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }

    response = requests.post(url, json=new_product, timeout=5)

    print("Response from server:\n")
    print(response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed.")

except requests.exceptions.Timeout:
    print("Request timed out.")

except Exception as e:
    print("Error:", e)
```

## Task 3 — Exception Handling

```python
# Function to safely divide two numbers

def safe_divide(a, b):

    try:
        result = a / b
        return result

    except ZeroDivisionError:
        return "Error: Cannot divide by zero"

    except TypeError:
        return "Error: Invalid input types"


# Testing the function
print("\nSafe Divide Tests:\n")

print(safe_divide(10, 2))     # normal case
print(safe_divide(10, 0))     # divide by zero
print(safe_divide("ten", 2))  # invalid type
```

```python
# Function to safely read a file

def read_file_safe(filename):

    try:
        with open(filename, "r", encoding="utf-8") as file:
            content = file.read()
            return content

    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")

    finally:
        print("File operation attempt complete.")


# Testing the function
print("\nReading existing file:\n")
print(read_file_safe("python_notes.txt"))

print("\nReading missing file:\n")
print(read_file_safe("ghost_file.txt"))
```

```python
def fetch_products():

    try:
        url = "https://dummyjson.com/products?limit=20"

        response = requests.get(url, timeout=5)

        return response.json()

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")

    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")

    except Exception as e:
        print("Error:", e)
```

```python
def create_product():

    try:
        url = "https://dummyjson.com/products/add"

        new_product = {
            "title": "My Custom Product",
            "price": 999,
            "category": "electronics",
            "description": "Created via API"
        }

        response = requests.post(url, json=new_product, timeout=5)

        return response.json()

    except requests.exceptions.ConnectionError:
        print("Connection failed.")

    except requests.exceptions.Timeout:
        print("Request timed out.")

    except Exception as e:
        print("Error:", e)


# Test
print("\nAPI Test:\n")
print(fetch_products())
print(create_product())
```

```python
print("\nProduct Lookup System:\n")

while True:

    user_input = input("Enter product ID (1-100) or 'quit': ")

    if user_input.lower() == "quit":
        break

    # Check if input is number
    if not user_input.isdigit():
        print("Invalid input. Please enter a number.")
        continue

    product_id = int(user_input)

    # Check valid range
    if product_id < 1 or product_id > 100:
        print("ID must be between 1 and 100.")
        continue

    try:
        url = f"https://dummyjson.com/products/{product_id}"

        response = requests.get(url, timeout=5)

        if response.status_code == 404:
            print("Product not found.")

        elif response.status_code == 200:
            data = response.json()
            print(f"{data['title']} - ${data['price']}")

    except requests.exceptions.ConnectionError:
        print("Connection failed.")

    except requests.exceptions.Timeout:
        print("Request timed out.")

    except Exception as e:
        print("Error:", e)
```

## Task 4 — Logging to File

```python
from datetime import datetime

# Function to log errors
def log_error(message):

    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    log_entry = f"[{timestamp}] ERROR: {message}\n"

    # Append to file
    with open("error_log.txt", "a", encoding="utf-8") as file:
        file.write(log_entry)
```

```python
import requests

print("\nTriggering connection error...\n")

try:
    # Invalid URL to simulate connection error
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)

except requests.exceptions.ConnectionError as e:
    print("Connection error occurred.")
    log_error(f"ConnectionError - {e}")
```

```python
print("\nTriggering 404 error...\n")

try:
    url = "https://dummyjson.com/products/999"  # invalid ID

    response = requests.get(url, timeout=5)

    if response.status_code != 200:
        print("Product not found.")
        log_error(f"HTTPError - 404 Not Found for product ID 999")

except requests.exceptions.ConnectionError:
    print("Connection failed.")

except requests.exceptions.Timeout:
    print("Request timed out.")

except Exception as e:
    print("Error:", e)
```
