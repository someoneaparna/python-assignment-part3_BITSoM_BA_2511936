```python

# TASK 1 — File Read & Write Basics


# Part A — Write the student notes to a file called python_notes.txt
# 'w' mode means we are creating the file (or overwriting if it exists)
# encoding="utf-8" ensures special characters are handled correctly

with open("python_notes.txt", "w", encoding="utf-8") as file:
    file.write("Topic 1: Variables store data. Python is dynamically typed.\n")
    file.write("Topic 2: Lists are ordered and mutable.\n")
    file.write("Topic 3: Dictionaries store key-value pairs.\n")
    file.write("Topic 4: Loops automate repetitive tasks.\n")
    file.write("Topic 5: Exception handling prevents crashes.\n")

# Print confirmation after writing
print("File written successfully.")

#-------------------------------------------------------------------------------

# Part A — Append two more lines to the same file
# 'a' mode means we are ADDING to the file without deleting existing content

with open("python_notes.txt", "a", encoding="utf-8") as file:
    file.write("Topic 6: Python uses indentation instead of curly braces.\n")
    file.write("Topic 7: Functions let you write code once and use it many times.\n")

# Print confirmation after appending
print("Lines appended.")


#-------------------------------------------------------------------------------

# Part B — Read the file back and print each line numbered
# 'r' mode means we are reading the file

with open("python_notes.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()  # reads all lines into a list

# Print each line numbered, stripping the \n at the end
for i, line in enumerate(lines, start=1):
    print(f"{i}. {line.strip()}")

# Count and print total number of lines
print(f"\nTotal lines in file: {len(lines)}")


#-------------------------------------------------------------------------------

# Ask the user for a keyword and print all matching lines
# case-insensitive means "Python" and "python" are treated the same

keyword = input("Enter a keyword to search: ").strip()

# Track if any matches were found
matches_found = False

for i, line in enumerate(lines, start=1):
    if keyword.lower() in line.lower():
        print(f"{i}. {line.strip()}")
        matches_found = True

# If no matches were found, print a helpful message
if not matches_found:
    print(f"No lines found containing '{keyword}'.")
```

```python
# =============================================================
# TASK 2 — API Integration
# =============================================================

# Install the requests library so we can make API calls
import subprocess
subprocess.run(["pip", "install", "requests"], capture_output=True)

# Import the requests library
import requests

print("requests library ready!")

#-------------------------------------------------------------------------------

# Step 1 — Fetch and Display Products
# Make a GET request to fetch 20 products from the API

url = "https://dummyjson.com/products?limit=20"

try:
    # Make the GET request with a 5 second timeout
    response = requests.get(url, timeout=5)

    # Parse the JSON response into a Python dictionary
    data = response.json()

    # The products are stored in the "products" key
    products = data["products"]

    # Print the table header
    print(f"{'ID':<4} | {'Title':<30} | {'Category':<14} | {'Price':<8} | {'Rating'}")
    print("-" * 70)

    # Loop through each product and print its details
    for product in products:
        print(f"{product['id']:<4} | {product['title']:<30} | {product['category']:<14} | ${product['price']:<8} | {product['rating']}")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"Unexpected error: {e}")


#-------------------------------------------------------------------------------

# Step 2 — Filter products with rating >= 4.5 and sort by price descending

# Filter only products with rating >= 4.5
filtered = [p for p in products if p["rating"] >= 4.5]

# Sort the filtered list by price in descending order (highest first)
filtered_sorted = sorted(filtered, key=lambda p: p["price"], reverse=True)

# Print the filtered and sorted list
print("\n--- Products with Rating ≥ 4.5 (sorted by price) ---")
print(f"{'ID':<4} | {'Title':<30} | {'Category':<14} | {'Price':<8} | {'Rating'}")
print("-" * 70)

for product in filtered_sorted:
    print(f"{product['id']:<4} | {product['title']:<30} | {product['category']:<14} | ${product['price']:<8} | {product['rating']}")


#-------------------------------------------------------------------------------

# Step 3 — Fetch all products in the laptops category

laptops_url = "https://dummyjson.com/products/category/laptops"

try:
    # Make the GET request
    laptops_response = requests.get(laptops_url, timeout=5)

    # Parse the JSON response
    laptops_data = laptops_response.json()
    laptops = laptops_data["products"]

    # Print each laptop's name and price
    print("\n--- Laptops ---")
    for laptop in laptops:
        print(f"{laptop['title']} — ${laptop['price']}")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"Unexpected error: {e}")

#-------------------------------------------------------------------------------

# Step 4 — Send a POST request to simulate creating a new product

post_url = "https://dummyjson.com/products/add"

# This is the data we are sending to the API
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

try:
    # Send the POST request with our data as JSON
    post_response = requests.post(post_url, json=new_product, timeout=5)

    # Print the full response from the server
    print("\n--- POST Request Response ---")
    print(post_response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"Unexpected error: {e}")
```

```python

# TASK 3 — Exception Handling


# Part A — Guarded Calculator

# This function safely divides two numbers and handles errors
def safe_divide(a, b):
    try:
        # Try to divide a by b
        return a / b
    except ZeroDivisionError:
        # This happens when b is 0
        return "Error: Cannot divide by zero"
    except TypeError:
        # This happens when a or b is not a number
        return "Error: Invalid input types"

# Test the function with three different cases
print("--- Safe Divide Tests ---")
print(f"safe_divide(10, 2)      = {safe_divide(10, 2)}")
print(f"safe_divide(10, 0)      = {safe_divide(10, 0)}")
print(f"safe_divide('ten', 2)   = {safe_divide('ten', 2)}")


#-------------------------------------------------------------------------------

# Part B — Guarded File Reader

# This function safely reads a file and handles errors
def read_file_safe(filename):
    try:
        # Try to open and read the file
        with open(filename, "r", encoding="utf-8") as file:
            content = file.read()
            return content
    except FileNotFoundError:
        # This happens when the file doesn't exist
        print(f"Error: File '{filename}' not found.")
    finally:
        # This always runs whether the file was found or not
        print("File operation attempt complete.")

# Test with a file that exists
print("\n--- Reading python_notes.txt ---")
content = read_file_safe("python_notes.txt")
if content:
    print(content)

# Test with a file that doesn't exist
print("--- Reading ghost_file.txt ---")
read_file_safe("ghost_file.txt")

#-------------------------------------------------------------------------------

# Part C — Robust API Calls
# All API requests are wrapped in try-except blocks

print("\n--- Part C: Robust API Call Test ---")

try:
    # Making a GET request with timeout
    test_response = requests.get("https://dummyjson.com/products/1", timeout=5)
    test_product = test_response.json()
    print(f"Product fetched: {test_product['title']} — ${test_product['price']}")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"Unexpected error: {e}")

#-------------------------------------------------------------------------------

# Part D — Input Validation Loop
# Keeps asking for a product ID until user types 'quit'

print("\n--- Part D: Product Lookup ---")

while True:
    # Ask the user for a product ID
    user_input = input("Enter a product ID to look up (1-100), or 'quit' to exit: ").strip()

    # Stop the loop if user types quit
    if user_input.lower() == "quit":
        print("Exiting product lookup.")
        break

    # Check if the input is a valid number
    if not user_input.isnumeric():
        print("⚠ Warning: Please enter a valid number between 1 and 100.")
        continue

    # Convert to integer and check the range
    product_id = int(user_input)
    if product_id < 1 or product_id > 100:
        print("⚠ Warning: Product ID must be between 1 and 100.")
        continue

    # Make the API call with the valid product ID
    try:
        lookup_response = requests.get(f"https://dummyjson.com/products/{product_id}", timeout=5)

        # Check the status code
        if lookup_response.status_code == 404:
            print("Product not found.")
        elif lookup_response.status_code == 200:
            lookup_product = lookup_response.json()
            print(f"✓ Product: {lookup_product['title']} — ${lookup_product['price']}")

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")
```

```python
# TASK 4 — Logging to File


# Import datetime to get the current timestamp
from datetime import datetime

# This function writes a timestamped error entry to error_log.txt
def log_error(location, error_type, message):
    # Get the current date and time
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # Format the log entry
    log_entry = f"[{timestamp}] ERROR in {location}: {error_type} — {message}\n"

    # Open the log file in append mode so logs accumulate
    with open("error_log.txt", "a", encoding="utf-8") as log_file:
        log_file.write(log_entry)

    # Also print it so we can see it in the output
    print(log_entry.strip())

print("--- Task 4: Error Logger ---\n")

#-------------------------------------------------------------------------------

# --- Trigger 1: ConnectionError ---
# Make a request to an unreachable URL to trigger a ConnectionError

print("Triggering ConnectionError...")
try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError:
    log_error("fetch_products", "ConnectionError", "No connection could be made")
except requests.exceptions.Timeout:
    log_error("fetch_products", "Timeout", "Request timed out")
except Exception as e:
    log_error("fetch_products", "UnexpectedError", str(e))



# --- Trigger 2: HTTP 404 Error ---
# Request a product ID that doesn't exist to trigger a 404 response

print("Triggering HTTP 404 error...")
try:
    bad_response = requests.get("https://dummyjson.com/products/999", timeout=5)

    # 404 is not a Python exception — we check status_code manually
    if bad_response.status_code != 200:
        log_error("lookup_product", "HTTPError", f"404 Not Found for product ID 999")

except requests.exceptions.ConnectionError:
    log_error("lookup_product", "ConnectionError", "No connection could be made")
except requests.exceptions.Timeout:
    log_error("lookup_product", "Timeout", "Request timed out")


# --- Read and print the full contents of error_log.txt ---

print("\n--- Contents of error_log.txt ---")

with open("error_log.txt", "r", encoding="utf-8") as log_file:
    print(log_file.read())


```
